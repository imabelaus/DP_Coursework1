### How much do variables weight?

```python
data=df.sort_values('Happiness.Rank',ascending=True)
df_wh=df.filter(['Country','Economy..GDP.per.Capita.','Family','Health..Life.Expectancy.','Freedom','Generosity','Trust..Government.Corruption.', "Dystopia.Residual"])
df_wh=df_wh.set_index('Country')
df_wh.plot.barh(stacked=True,  figsize=(10,28))
plt.title("Contribution of Factors")
plt.show()
```

```python
gender1 = ('Distopia Residual',"Family", 'GDP per Capita', "Life Expectancy", "Freedom", "Generosity", "Trust in Government",  )
y_pos = np.arange(len(gender1))
distribution = [1.85,1.189,.985,.55,.409,.247, .1231]
 
plt.bar(y_pos, distribution, alpha=.9)
plt.xticks(y_pos, gender1, rotation='vertical')
plt.ylabel('mean')
plt.title('Mean of Variables')
 
plt.show()
```

### Reduced dataset to see correlation among variables 

```python
corrdf = df.loc[:,['Happiness.Score','Economy..GDP.per.Capita.','Family','Health..Life.Expectancy.', "Freedom", "Generosity", "	Trust..Government.Corruption."]]
```

```python
plt.figure(figsize=(16,10))  # on this line I just set the size of figure to 12 by 10.
sns.heatmap(corrdf.corr(), annot=True, linewidths=1, cmap="ocean_r", fmt=".2f")  # seaborn has very simple solution for heatmap

plt.suptitle("Correlation Map", fontsize=18)

plt.show()
```

### How is money related to happiness?

```python
sns.lmplot(y = "Happiness.Score", x="Economy..GDP.per.Capita.", data=df, size=5,  palette="Set2")
plt.show()
```

### Region Analysis

```python
countries = ('Sub-Saharan Africa',"Central and Eastern Europe", 'Latin America and Caribbean', "Western Europe", "Middle East and Northern Africa", "Southeastern Asia", "Southern Asia", "Eastern Asia", "Australia and New Zealand", "North America")
y_pos = np.arange(len(countries))
distribution = [40,29,22,21,19,8,6,6,2,2]

plt.bar(y_pos, distribution, alpha=.9)
plt.xticks(y_pos, countries, rotation='vertical')
plt.ylabel('quantity')
plt.title('Number of countries per Region')
 
plt.show()
```
